```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sunrise.jpg', 1)
```

```python
plt.imshow(img)
plt.show()
```

```python
cv2.imshow('img', img)
plt.figure('Hist')
plt.subplot(2, 1, 1)
plt.hist(img.ravel(), bins=256, range=[0, 256])
plt.show()
k = cv2.waitKey(0)  # waitkey代表读取键盘的输入，括号里的数字代表等待多长时间，单位ms。 0代表一直等待
if k == 27:  # 键盘上Esc键的键值
    cv2.destroyAllWindows()
```

```python
im_equ = cv2.equalizeHist(img)
cv2.imshow("equ1", im_equ)

plt.subplot(2, 1, 2)
plt.hist(im_equ.ravel(), bins=256, range=[0, 256])
plt.show()
cv2.waitKey()  # waitkey代表读取键盘的输入，括号里的数字代表等待多长时间，单位ms。 0代表一直等待
cv2.destroyAllWindows()
```
